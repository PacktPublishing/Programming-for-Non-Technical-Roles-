```python
import pandas as pd

company_df = pd.read_csv("full company data.csv", parse_dates=["Date"], index_col="Date")

competitor_df = pd.read_csv("full competitor data.csv", parse_dates=["Date"], index_col="Date")

market_df = pd.read_csv("market data.csv", parse_dates=["Date"], index_col="Date")


company_df


```

```python
company_df.mean().round(2)
```

```python
company_df.describe().round(2)
```

```python
company_df["Market Share pc"] = company_df["Revenue"] / market_df["Revenue"] * 100
company_df
```

```python
company_df["Market Share pc"].max()
```

```python
company_df["Percent Change"] = (company_df["Market Share pc"] - company_df["Market Share pc"][0]) / company_df["Market Share pc"][0] *100
company_df
```

```python
#check if there is correlation between the add spent and revenue
company_df["Ad Spent"].corr(company_df["Percent Change"])
```
