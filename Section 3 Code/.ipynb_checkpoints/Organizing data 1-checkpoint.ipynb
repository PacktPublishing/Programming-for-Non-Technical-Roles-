```python
import pandas as pd

company_df = pd.read_csv("company financials.csv", names=["Date","Revenue","NET Income","Ad Spent","Products launched","Profit Margins"], header=0)

format = "%Y-%m-%d"

company_df['Date'] = pd.to_datetime(company_df['Date'], format=format)

company_df.set_index("Date", inplace=True)

company_df
```

```python
competitor_df = pd.read_csv("competitor financials.csv", names=[
    "Date", "Revenue", "NET Income", "Ad Spent", "Products launched", "Profit Margins"], header=0)

format = "%Y-%m-%d"

competitor_df['Date'] = pd.to_datetime(competitor_df['Date'], format=format)

competitor_df.set_index("Date", inplace=True)

competitor_df
```
