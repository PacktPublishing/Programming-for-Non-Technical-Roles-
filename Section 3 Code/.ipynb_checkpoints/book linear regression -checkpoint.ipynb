```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import style
style.use('seaborn-poster')
import mpld3
mpld3.enable_notebook()
import pandas as pd
import sklearn.linear_model
import numpy as np

company_df = pd.read_csv("fully complete company data.csv", parse_dates=["Date"], index_col="Date")


X = company_df["Ad Spent"].values.reshape(-1, 1)
Y = company_df["Revenue"].values.reshape(-1, 1)

data = company_df[["Ad Spent", "Revenue"]]

data.plot(kind="scatter", x="Ad Spent", y="Revenue")

model = sklearn.linear_model.LinearRegression()

model.fit(X, Y)

X_rev = [[200000]]

print(model.predict(X_rev))
```
