```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import style
style.use('seaborn-poster')
import mpld3
mpld3.enable_notebook()
import pandas as pd


company_df = pd.read_csv("fully complete company data.csv", parse_dates=["Date"], index_col="Date")

competitor_df = pd.read_csv("fully complete competitor data.csv", parse_dates=["Date"], index_col="Date")

market_df = pd.read_csv("market data.csv", parse_dates=["Date"], index_col="Date")

company_df

```

```python
company_market_share = company_df["Market Share pc"].plot()
company_market_share.set_ylabel("Market Share Percentage")
```

```python
print(plt.style.available)
```

```python
company_percentage_change = company_df["Percent Change"].plot()
company_percentage_change.set_ylabel("Percent change")
```

```python
profit_margins = company_df["Profit Margins"].plot()
profit_margins.set_ylabel("Profit Margins")
```

```python
# company_df["Market Share pc"] = company_df["Market Share pc"] * 10000
# company_df["Percent Change"] = company_df["Percent Change"] * 100000
company_plot = company_df[["Revenue", "NET Income", "Ad Spent", "Profit Margins", "Market Share pc", "Percent Change"]].plot()
company_plot
```

```python
company_change = company_df[["Market Share pc", "Percent Change"]].plot()
company_change
```

```python
# company_df["Percent Change"] = company_df["Percent Change"] * 10000
ad_spent_corr = company_df[["Ad Spent","Percent Change"]].plot()
ad_spent_corr
```

```python
# company_df["Profit Margins"] = company_df["Profit Margins"] * 100
company_profit_market = company_df[["Profit Margins", "Market Share pc"]].plot()
company_profit_market
```

```python
revenue_corr = company_df[["Ad Spent","Revenue"]].plot()
revenue_corr
```

```python
market_rev_corr = market_df[["Revenue", "Ad Spent"]].plot()
market_rev_corr
```

```python
#save graphs as images 
fig1 = market_rev_corr.get_figure()
fig1.savefig("market_rev_corr.png")

fig2 = revenue_corr.get_figure()
fig2.savefig("company_revenue_corr.png")

fig3 = company_profit_market.get_figure()
fig3.savefig("company_profit_market.png")

fig4 = ad_spent_corr.get_figure()
fig4.savefig("company_ad_spent_corr.png")

fig5 = company_change.get_figure()
fig5.savefig("company_change.png")

fig6 = company_plot.get_figure()
fig6.savefig("company_plot.png")
```
