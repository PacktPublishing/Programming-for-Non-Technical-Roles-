```python
import pandas as pd
%matplotlib inline

company_df = pd.read_csv("full company data.csv", parse_dates=["Date"], index_col="Date")
company_df
```

```python
competitor_df = pd.read_csv("full competitor data.csv", parse_dates=["Date"], index_col="Date")
competitor_df
```

```python
competitor_df = competitor_df.interpolate(method="values")
# competitor_df = competitor_df.fillna(method= "bfill")
competitor_df
```

```python
competitor_df = competitor_df.round(2)
competitor_df
```

```python
market = company_df.add(competitor_df[[
    "Ad Spent","Products launched", "Revenue", "NET Income"]])

market_df = market[[
    "Revenue", "NET Income","Ad Spent","Products launched"]]

market_df
```

```python
market_df.plot()
```

```python
market_df.to_csv("market data.csv")

competitor_df.to_csv("full competitor data.csv")
```
